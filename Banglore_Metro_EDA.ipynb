```python
import pandas as pd

# Load the dataset
file_path = 'NammaMetro_Ridership_Dataset.csv'
metro_data = pd.read_csv('NammaMetro_Ridership_Dataset.csv')

# Convert "Record Date" to datetime format
metro_data['Record Date'] = pd.to_datetime(metro_data['Record Date'], format='%d-%m-%Y', errors='coerce')

# Extract the month from the "Record Date"
metro_data['Month'] = metro_data['Record Date'].dt.month

# Display the first few rows to confirm
print(metro_data[['Record Date', 'Month']].head())
```

**How does the total ridership (e.g., "Total Smart Cards" + "Tokens" + "Total QR") vary over time?

```python
metro_data['Month'].unique()
```

```python
# Define a mapping dictionary
month_mapping = {
    10: 'October',
    11: 'November',
    12: 'December',
    1: 'January'
}

# Map the "Month" column to month names
metro_data['Month Name'] = metro_data['Month'].map(month_mapping)

# Display the updated DataFrame to confirm
print(metro_data[['Record Date', 'Month', 'Month Name']].head())
```

```python
df=pd.read_csv('NammaMetro_Ridership_Dataset.csv')
# Assume 'df' is the intended DataFrame, replacing 'metro' and 'metro_data' with 'df'
df['Record Date'] = pd.to_datetime(df['Record Date'], format='%d-%m-%Y')
df['total_ridership'] = df[['Total Smart Cards', 'Tokens', 'Total QR']].sum(axis=1)
plt.figure(figsize=(12, 6))
plt.plot(df['Record Date'], df["total_ridership"], marker='o')
plt.title("total ridership over time")
plt.xlabel('date')
plt.ylabel('total ridership')
plt.grid()
plt.show()
```

**Trends in Ticket Types Over Time

```python
ticket_types = ['Stored Value Card', 'One Day Pass', 'Three Day Pass', 'Five Day Pass', 'Tokens']
plt.figure(figsize=(12, 6))

# Replace 'metro_data' with 'df' to use the correct DataFrame
for ticket in ticket_types:
    plt.plot(df['Record Date'], df[ticket], label=ticket)

plt.title('Trends in Ticket Types Over Time')
plt.xlabel('Date')
plt.ylabel('Count')
plt.legend()
plt.grid()
plt.show()
```
